```python
def grafico_violino(valores, inicio, fim):
    dados_plot = pd.concat([diagnostico, valores.iloc[:, inicio:fim]], axis=1)
    dados_plot = pd.melt(dados_plot, id_vars="diagnostico", var_name="exames", value_name="valores")
    ax = sns.violinplot(x=dados_plot.exames, y=dados_plot.valores, hue=dados_plot.diagnostico, split=True)
    ax.figure.set_size_inches(10, 10)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    ax

grafico_violino(valores_exames_v2, 10, 21)
```

```python
valores_exames_v3 = valores_exames_v2.drop(columns=["exame_29", "exame_4"])

def classificar(valores):
    SEED = 1234
    np.random.seed(SEED)

    X_treino, X_teste, y_treino, y_teste = train_test_split(valores, diagnostico, test_size=0.3)

    classificador = RandomForestClassifier()
    classificador.fit(X_treino, y_treino)
    print(f"Resultado da classificação: {classificador.score(X_teste, y_teste) * 100:0.2f}%")

classificar(valores_exames_v3)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

selecionar_kmelhores = SelectKBest(chi2, k=5)
```

```python
valores_exames_v6 = valores_exames_v1.drop(columns=["exame_4", "exame_29", "exame_3", "exame_24"])
```

```python
SEED = 1234
np.random.seed(SEED)

X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

treino_kbest = selecionar_kmelhores.fit(X_treino, y_treino)
treino_kbest = selecionar_kmelhores.transform(X_treino)
teste_kbest = selecionar_kmelhores.transform(X_teste)
```

```python
treino_kbest.shape
```

```python
classificador = RandomForestClassifier(random_state=1234)
classificador.fit(treino_kbest, y_treino)
print(f"Resultado da classificação: {classificador.score(teste_kbest, y_teste) * 100:0.2f}%")
```

```python
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(y_teste, classificador.predict(teste_kbest))
```

```python
matriz_confusao
```

```python
sns.set(font_scale=2)
ax = sns.heatmap(matriz_confusao, annot=True, fmt="d") # d = inteiro
ax.figure.set_size_inches(10, 8)
ax.set_xlabel("Predição", fontsize=14)
ax.set_ylabel("Real", fontsize=14)
ax
```

```python
from sklearn.feature_selection import RFE

SEED = 1234
np.random.seed(SEED)

X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

classificador = RandomForestClassifier(random_state=1234)
classificador.fit(X_treino, y_treino)

selecionador_rfe = RFE(estimator=classificador, n_features_to_select=5, step=1)
selecionador_rfe.fit(X_treino, y_treino)
treino_rfe = selecionador_rfe.transform(X_treino)
teste_rfe = selecionador_rfe.transform(X_teste)

classificador.fit(treino_rfe, y_treino)

matriz_confusao = confusion_matrix(y_teste, classificador.predict(teste_rfe))
sns.set(font_scale=2)
ax = sns.heatmap(matriz_confusao, annot=True, fmt="d") # d = inteiro
ax.figure.set_size_inches(10, 8)
ax.set_xlabel("Predição", fontsize=14)
ax.set_ylabel("Real", fontsize=14)
ax

print(f"Resultado da classificação: {classificador.score(teste_rfe, y_teste) * 100:0.2f}%")
```

```python
from sklearn.feature_selection import RFECV

SEED = 1234
np.random.seed(SEED)

X_treino, X_teste, y_treino, y_teste = train_test_split(valores_exames_v6, diagnostico, test_size=0.3)

classificador = RandomForestClassifier(random_state=1234)
classificador.fit(X_treino, y_treino)

selecionador_rfecv = RFECV(estimator=classificador, cv=5, step=1, scoring="accuracy", n_jobs=-1)
selecionador_rfecv.fit(X_treino, y_treino)
treino_rfecv = selecionador_rfecv.transform(X_treino)
teste_rfecv = selecionador_rfecv.transform(X_teste)

classificador.fit(treino_rfecv, y_treino)

matriz_confusao = confusion_matrix(y_teste, classificador.predict(teste_rfecv))
sns.set(font_scale=2)
ax = sns.heatmap(matriz_confusao, annot=True, fmt="d") # d = inteiro
ax.figure.set_size_inches(10, 8)
ax.set_xlabel("Predição", fontsize=14)
ax.set_ylabel("Real", fontsize=14)
ax

print(f"Resultado da classificação: {classificador.score(teste_rfecv, y_teste) * 100:0.2f}%")
```
